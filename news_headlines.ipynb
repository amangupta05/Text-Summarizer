```python
from transformers import AutoTokenizer, DataCollatorForSeq2Seq, AutoModelForSeq2SeqLM, TrainingArguments, Trainer
from datasets import Dataset, DatasetDict
#import tensorflow as tf
import pandas as pd
#import keras
from sklearn.model_selection import train_test_split
#import keras_nlp
```

```python
dataset = pd.read_csv('Articles.csv', encoding='Windows-1252')
dataset.drop(columns=['Date', 'NewsType'], inplace=True)
dataset['Heading'] = dataset['Heading'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x).astype(str)
```

```python
train_set, test_set = train_test_split(dataset, test_size=0.2, random_state=42)
train_set, validation_set = train_test_split(train_set, test_size=0.25, random_state=42)
```

```python
train_dataset = Dataset.from_pandas(train_set)
validation_dataset = Dataset.from_pandas(validation_set)
test_dataset = Dataset.from_pandas(test_set)
```

```python
model_path = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
```

```python
train_encoding = tokenizer(train_dataset['Article'], truncation=True, padding=True, return_tensors='pt')
validation_encoding = tokenizer(validation_dataset['Article'], truncation=True, padding=True, return_tensors='pt')
test_encoding = tokenizer(test_dataset['Article'], truncation=True, padding=True, return_tensors='pt')
```

```python
def tokenization(examples):
    inputs = ['Heading: ' + doc for doc in examples["Article"]]
    tok = tokenizer(inputs, truncation=True, padding = True, return_tensors='pt',max_length=1024)


    labels = tokenizer(examples['Heading'],truncation=True, padding = True, return_tensors='pt',max_length=128)

    tok['labels'] = labels['input_ids']

    return {'input_ids': tok['input_ids'].to(device),
            'labels': tok['labels'].to(device),
            'attention_mask': tok['attention_mask'].to(device)}
```

```python
train_dataset = Dataset.from_dict({
    'input_ids': train_encoding['input_ids'],
    'attention_mask': train_encoding['attention_mask'],
    'Article': train_dataset['Article'],  # Include the 'Article' column
    'Heading': train_dataset['Heading']
})

validation_dataset = Dataset.from_dict({
    'input_ids': validation_encoding['input_ids'],
    'attention_mask': validation_encoding['attention_mask'],
    'Article': validation_dataset['Article'],  # Include the 'Article' column
    'Heading': validation_dataset['Heading']
})

test_dataset = Dataset.from_dict({
    'input_ids': test_encoding['input_ids'],
    'attention_mask': test_encoding['attention_mask'],
    'Article': test_dataset['Article'],  # Include the 'Article' column
    'Heading': test_dataset['Heading']
})

# Create DatasetDict
dataset_dict = DatasetDict({
    'train': train_dataset,
    'validation': validation_dataset,
    'test': test_dataset
})
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import os

folder_path = "./results"
items = os.listdir(folder_path)

subfolders = [item for item in items if os.path.isdir(os.path.join(folder_path, item))]

if subfolders:
  first_model = subfolders[0]

# Load the trained summarization model
model_name = f"./results/{first_model}"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
# dataset = pd.read_csv('Articles.csv', encoding='Windows-1252')
dataset = pd.DataFrame(tokenized_dataset['test'])
dataset['Article'] = dataset['Article'].astype(str)
```

```python
# Function to generate summaries
def generate_summary(text):
    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)
    outputs = model.generate(inputs["input_ids"], max_length=150, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return summary
```

```python
# Apply the function to each row in the dataset
dataset['Summary'] = dataset['Article'].apply(generate_summary)
```

```python
# Save the summarized dataset to a new CSV file
dataset.to_csv('Summarized_Articles.csv', index=False)
```
